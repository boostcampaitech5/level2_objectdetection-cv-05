```python
import random

def random_combs(repeat, num, out):
    total=[]
    for i in range(repeat):  # 3
        nums = set(range(num))  # 127 -> 0부터 126까지의 정수를 set으로 생성
        combos = set()  # 추출된 3개의 숫자의 조합을 담을 set
        try:
            while nums:
                combo = random.sample(nums, out)  # 3개의 숫자를 뽑아서 조합 생성
                combos.add(tuple(sorted(list(combo))))  # 추출된 조합을 set에 추가
                nums -= set(combo)  # 추출된 숫자들을 전체 집합에서 제거
        except:
            pass
        print("총", len(combos), "개의 조합 combo", i+1 ,"을 생성했습니다.")
        
        total.append(list(combos))

    result = set(item for sublist in total for item in sublist)
    combi = list(result)
    return combi

ratio_2_2_metal = random_combs(3,cnt0,2)  ### 총 필요한 class image개수/보유 image개수, 보유 image개수, 뽑을 숫자 수
ratio_2_2_glass = random_combs(3,cnt1,2)
```

```python
# 조합별로 이미지 불러와서 노이즈이미지에 추가
# 1) 128 < h, w < 384
# 2) h:w -> 0.9 ~ 1.1
# 3)
# x, y -> (128,128), (512,128), (128,512), (512,512)
# A -> 128,128 < x, y < 512-w,512-h
# B -> 512,128 < x, y < 896-w,512-h
# C -> 512,512 < x, y < 896-w,896-h
# D -> 128,512 < x, y < 896-w,896-h
```

```python
def locate_cut_img(img_loc, ratio):
    h, w = 0, 0
    if ratio < 0.25:  # 가로에 비해 세로가 너무 작은 경우
        ratio = 0.25
    elif ratio > 4:   # 가로에 비해 세로가 너무 큰 경우
        ratio = 4
        
    while not (90 <= w < 384 and 90 <= h < 384):
        w = random.randint(90, 384)
        h = int(w*ratio)

    if img_loc == 'A':
        # A의 x, y를 랜덤하게 선택합니다.
        x = random.randint(128, 512-w)
        y = random.randint(128, 512-h)
    elif img_loc == 'B':
        # B의 x, y를 랜덤하게 선택합니다.
        x = random.randint(512, 896-w)
        y = random.randint(128, 512-h)
    elif img_loc == 'C':
        # C의 x, y를 랜덤하게 선택합니다.
        x = random.randint(512, 896-w)
        y = random.randint(512, 896-h)
    elif img_loc == 'D':
        # D의 x, y를 랜덤하게 선택합니다.
        x = random.randint(128, 512-w)
        y = random.randint(512, 896-h)

    return (w,h,x,y)
```

```python
def mk_anno(anno_list, img_class, img_name, bbox):
    global id
    anno_dict = {
        'area': bbox[2]*bbox[3],
        'bbox': bbox,
        'category_id': label_name.index(img_class),
        'id': id,
        'image_id':img_name,
        'iscrowd': 0
        }
    anno_list.append(anno_dict)
```

```python
def paste_image_2_2(save_dir,noise_img, img_class, img_ids, img_name, anno_list):  # n:4883 부터 시작
    global id
    for i in range(len(img_ids)):
        if img_class=='Metal':
            paths = 'Metal_'+str(img_ids[i]).zfill(4)+'.jpg'  # 사진 불러오기
            imgsize = Image.open('/opt/ml/mosaic_glassmetal/'+paths).size   ### 이미지 불러올 주소
            ratio = imgsize[1]/imgsize[0]  # 비율구하기 h/w
            if i == 0:
                w, h, x, y = locate_cut_img('C',ratio)
            if i == 1:
                w, h, x, y = locate_cut_img('D',ratio)
            
        elif img_class=='Glass':
            paths = 'Glass_'+str(img_ids[i]).zfill(4)+'.jpg'  # 사진 불러오기
            imgsize = Image.open('/opt/ml/mosaic_glassmetal/'+paths).size   ### 이미지 불러올 주소
            ratio = imgsize[1]/imgsize[0]  # 비율구하기
            if i == 0:
                w, h, x, y = locate_cut_img('A',ratio)  # 붙일 이미지 크기, 좌표 저장
            elif i == 1:
                w, h, x, y = locate_cut_img('B',ratio)

        image_path = '/opt/ml/mosaic_glassmetal/'+paths
        resized_img = Image.open(image_path).resize((w,h))  # 이미지 size 변경

        bbox = [x,y,w,h]
        mk_anno(anno_list,img_class, img_name, bbox)
        id+=1
        
        noise_img.paste(resized_img, (int(bbox[0]), int(bbox[1])))  # x,y min좌표에 이미지 붙임
        
    noise_img.save(save_dir+str(img_name).zfill(4)+'.jpg')  ### 4컷 image 저장할 디렉터리 설정 / 0000.jpg로 파일 저장
```

```python
import time
import numpy as np

width = 1024
height = 1024
# num_of_images = int(num_of_images)
mosaic_save_dir='/opt/ml/dataset/train/'
# if not os.path.exists(mosaic_save_dir):
#     os.makedirs(mosaic_save_dir)

id = 24400 # 24399 +1부터 시작
anno_list = []
images_list = []
img_name=5197
for n in range(len(ratio_2_2_metal)):

    rgb_array = np.random.rand(height,width,3) * 255
    noise_image = Image.fromarray(rgb_array.astype('uint8')).convert('RGB')
    
    print(ratio_2_2_metal[n])
    paste_image_2_2(mosaic_save_dir, noise_image, 'Glass', ratio_2_2_glass[n], img_name, anno_list)  # general 이미지 붙이기
    paste_image_2_2(mosaic_save_dir, noise_image, 'Metal', ratio_2_2_metal[n], img_name, anno_list)  # cloth image 붙이기
    
    images_list.append({
    'width': 1024,
    'height': 1024,
    'file_name': 'train/'+str(img_name).zfill(4)+'.jpg',
    'license': 0,
    'flickr_url': None,
    'coco_url': None,
    'date_captured': '2020-12-23 16:20:30',
    'id': img_name
    })
    
    img_name+=1
    print(n)
        
print('done')

# else:
#     print('file alreay exist, change save_dir or data will be overwritten')
```
